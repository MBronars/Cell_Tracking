```python
import skimage 
import cv2 as cv
from skimage import data, io, filters, measure
from matplotlib import pyplot as plt
from skimage.transform import rescale, resize
from PIL import Image
import PIL
import numpy as np
from PIL import ImageSequence
from overlap import getOverlap
import math 
from cellClass import Cell

```

```python
overlaps = getOverlap('Test.tif','protein.png',False)
def distance(x,y):  
     dist = math.sqrt((x[0][0] - y[0][0])**2 + (x[0][1] - y[0][1])**2)  
     return dist 
print(overlaps)
```

```python
#Initialize a list that will eventually hold the overlap data for each cell as it travels through the stack
cells = []
curr = 0
count = 0
name = 65
index = -1
while index < len(overlaps)-1:
    index += 1
    #print("loop")
    x = overlaps[index]
    if x[2] == curr:
        #print("if")
        if x[2] == 0:
            #print("0")
            pos = x[0]
            c = Cell(chr(name),0,pos)
            c.used = True
            cells.append(c)
            name += 1
            continue
        best = 999999
        dex = 0
        i = 0
        for c in cells:
            if c.used or c.gone:
                #print("used")
                i+=1
                continue
            #print("not used")
            dist = distance([c.last],x)
            if dist < best:
                best = dist
                dex = i
            i += 1
        #print("post used")
        #print("dex = " + str(dex))
        cell = cells[dex]
        cell.history.append(x[0])
        cell.used = True
        if best == 999999:
            #print("new")
            n = "New" + chr(name)
            c = Cell(n, curr, x[0])
            c.used = True
            cells.append(c)
            name += 1
    else:
        #print("else")
        for c in cells:
            if c.used:
                #print("reset")
                c.used = False
            elif c.gone:
                #print("gone")
                continue
            else:
                #print("now gone")
                c.gone = True
        curr += 1
        name = 65
        index -= 1
```

```python
for cell in cells:
    cell.displayData()
```
